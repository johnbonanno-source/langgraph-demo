```python
from typing import List, TypedDict, Optional
from langgraph.graph import StateGraph
```

```python
# We now create an AgentState - shared data structure that keeps track of information as your application runs. 

class AgentState(TypedDict): # Our state schema
    values: List[int]
    name : str
    operation: str
    result: Optional[int]
```

```python
def process_operator(state: AgentState) -> AgentState:
    """This function handles a list of inputs and decides what to do with them based on the operator"""

    if state['operation'] == '+':
        ans = 0
        for element in state['values']:
            ans += element
        state['result'] = ans
    elif state['operation'] == '*':
        ans = 1
        for element in state['values']:
            ans *= element
        state['result'] = ans
    return state
    
    
```

```python
graph = StateGraph(AgentState)
graph.add_node("handle_operators",process_operator)
graph.set_entry_point("handle_operators")
graph.set_finish_point("handle_operators")
app = graph.compile()
```

```python
answers = app.invoke({"name":'Elements_to_process',"values": [1,3,3],"operation":'+'})
```
